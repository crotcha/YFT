```r
#Bartlett’s sphericity test: Tests the hypothesis that correlations between variables are greater 
#than would beexpected by chance: Technically, tests if the matrix is an identity matrix. 

#The p-value should be significant:i.e., the null hypothesis that all off-diagonal correlations are zero is falsified.

#Prof. Shigenobu Aoki provides the following function to conduct Bartlett’s sphericity test at his web page:

Bartlett.sphericity.test <- function(x){
    method <- "Bartlett’s test of sphericity"
    data.name <- deparse(substitute(x))
    x <- subset(x, complete.cases(x)) # Omit missing values
    n <- nrow(x)
    p <- ncol(x)
    chisq <- (1-n+(2*p+5)/6)*log(det(cor(x)))
    df <- p*(p-1)/2
    p.value <- pchisq(chisq, df, lower.tail=FALSE)
    names(chisq) <- "X-squared"
    names(df) <- "df"
    return(structure(list(statistic=chisq, parameter=df, p.value=p.value,method=method, data.name=data.name), class="htest"))
}
```

# Teste de Kaiser-Meyer-Olkin e Bartlett de adequabilidade da matriz de correlações
Analise-se agora a adequabilidade da matriz de correlações através da Estatística Kaiser-
Meyer-Olkin e do Teste de Bartlett:

O Teste de Bartlet permite verificar se a matriz de correlações é significativamente
diferente da matriz identidade. 

As Hipóteses em teste são: H0: A matriz de correlações é a
matriz identidade (o que isto significa é que as correlações entre as variáveis são zero) vs
H1: A matriz de correlações não é a matriz identidade. Pretende-se rejeitar H0 para poder
fazer uma ACP.

Analisando a significância do teste de Bartlett, pode inferir-se que as variáveis estão
correlacionadas entre si (Sig=0,000<0,05 logo rejeita-se H0).

Por outro lado, sendo o KMO=0,84, a adequação da ACP aos dados é Boa (ver tabela
de classificação da ACP a seguir); Este valor não desaconselha a utilização de uma ACP a

# Aplicação do Algoritmo CLARA (Clustering Large Applications)


Devido a limitação de memória do algoritmo k-means recorreu-se ao algoritmo CLARA(Clustering Large Applications, Kaufman and Rousseeuw (1990)).

Este algoritmo é apresentado como uma extensão do algoritmo PAM(K-Medoids), que recorre a uma abordagem de sampling para reduzir o tempo de execução e problemas RAM.

A literatura refere que o k-Medoids é mais robusta que o k-means, menos sensível a outliers, por utilizar medoids em vez da média como centroids.


As variáveis de agrupamento utilizadas nos passos seguintes são as 4 componentes principais.

## Avaliar a existência de Clusters nos dados

Uma das fragilidades dos algoritmos de clustering é que devolvem sempre grupos, mesmo tratando-se apenas de grupos aleatórios. Por esse motivo, antes de aplicar o algoritmo de cluster, é importante avaliar se os padrões existentes nos dados não se tratam de estruturas aleatórias.

##### Para avaliar a tendência de Clusters nos dados, utiliza-se a abordagem VAT(visual assessment of cluster tendency)(Bezdekand Hathaway, 2002):

Calcula-se uma matriz de dissimilaridade ordenada. O grau de dissimilaridade entre casos é representado como uma distância entre pontos projetados no espaço multivariado.

Na figura as cores devem ser interpretadas da seguinte forma:<br/><br/>
<b>Vermelho</b>: Alta similaridade(i.e baixa dissimilaridade)<br/>
<b>Azul</b>: Baixa similaridade<br/>

Pela analise do gráfico podemos confirmar que existem grupos não aleatórios nos dados. 


```r
#get_dist(comp, method = "pearson")
fviz_dist(get_dist(sample_frac(comp, 0.05), method = "euclidean"), show_labels=FALSE)
```

## Determinar o número óptimo de Clusters

Como iremos utilizar um algoritmo de particionamento onde o número de clusters K é um parâmetro obrigatório, é necessário determinar o valor óptimo para k.

A decisão do número de clusters que melhor representa uma solução é crucial para o resultado final da análise. Embora a discussão de como determinar o número de cluster remonta o trabalho de Thorndike (1953), nenhuma solução satisfatória para o problema foi encontrada até hoje, apenas são sugeridos um conjunto de euristicas e indices, (Donlicar 2002), analisados e comparados por Milligan (1981), Milligan e Cooper (1985). 

Seguidamente recorre-se a métodos diretos e estatisticos para determinar um possivel valor de k:<br/>

1 - Métodos diretos: Consiste em critérios de optimização, como por exemplo o coeficiente da silhueta ou o método do cotovelo(elbow)<br/><br/>
2 - Métodos de testes estatisticos: Consiste em comparar evidências com a hipotese nula. Por exemplo o método de estatistica GAP. 

Os resultado encontrados pela aplicação de diferentes metodos apresentam valores entre 5 e 8 clusters.
Iremos testar o algoritmo com K entre 4 e 8 Clusters e selecionar o resultado com melhor coeficiente de silhueta.


```r
comp1 <- data.frame(pca1$x[,1:4])
#Elbow method
fviz_nbclust(comp1, kmeans, method = "wss") +
  geom_vline(xintercept = 5, linetype = 2)+
  labs(subtitle = "Elbow method")

# Silhouette method
fviz_nbclust(comp1, kmeans, method = "silhouette")+
  labs(subtitle = "Silhouette method")
# Gap statistic
# nboot = 50 to keep the function speedy.
# recommended value: nboot= 500 for your analysis.
# Use verbose = FALSE to hide computing progression.
#set.seed(123)

fviz_nbclust(comp1, kmeans, nstart = 25, method = "gap_stat", nboot = 50)+
  labs(subtitle = "Gap statistic method")

#nb <-NbClust(comp1,distance ="euclidean",min.nc =2,max.nc =10,method ="kmeans")
#fviz_nbclust(nb)+
#labs(subtitle = "Gap statistic method")
```

### Que medidas de distância escolher ?

A escolha da medida de distância é de extrema importância, influenciando o resultado final da análise. 
È usual, por omissão, os algoritmos utilizarem a distância <b>Euclidiana e de Manhattan</b>, mas dependendo dos dados em análise e das questões de investigação podem ser utilizadas outras medidas de dissimiliridade como os <b>coeficientes de correlação</b>, descritos como medidas de forma por serem insensíveis às diferenças de magnitude das variáveis envolvidas no cálculo dos coeficientes, (Aldenderfer e Blashfield 1984; Kassambara 2017). 

Uma vez que pretendemos agrupar os objetos de acordo com a magnitude da pontuação das respostas do questionário, foram consideradas as distâncias Euclideanas e de Manhattan.
 
Foi escolhida a distância Euclideana porque o resultado em termos de coeficiente de silhueta foi melhor. 

Foram testados valores de K=4,5,6,7 e 8.

k=6 apresenta o melhor coeficiente de silhueta, 0,8 .

```r
clara.res_k4 <- clara(comp, 4, samples = 50, pamLike = TRUE, metric='euclidean')
fviz_cluster(clara.res_1, data = comp, stand = FALSE,
             ellipse = FALSE, show.clust.cent = FALSE,
             geom = "point",palette = "jco", ggtheme = theme_classic())
fviz_silhouette(clara.res_k4, palette = "jco", ggtheme = theme_classic())

clara.res_k5 <- clara(comp, 5, samples = 50, pamLike = TRUE, metric='euclidean')
fviz_cluster(clara.res_2, data = comp, stand = FALSE,
             ellipse = FALSE, show.clust.cent = FALSE,
             geom = "point",palette = "jco", ggtheme = theme_classic())
fviz_silhouette(clara.res_k5, palette = "jco", ggtheme = theme_classic())


clara.res <- clara(comp, 6, samples = 50, pamLike = TRUE,metric='euclidean')
fviz_cluster(clara.res, data = comp, stand = FALSE,
             ellipse = FALSE, show.clust.cent = FALSE,
             geom = "point",palette = "jco", ggtheme = theme_classic())
fviz_silhouette(clara.res, palette = "jco", ggtheme = theme_classic())


clara.res_k7 <- clara(comp, 7, samples = 50, pamLike = TRUE,metric='euclidean')
fviz_cluster(clara.res, data = comp, stand = FALSE,
             ellipse = FALSE, show.clust.cent = FALSE,
             geom = "point",palette = "jco", ggtheme = theme_classic())
fviz_silhouette(clara.res_k7, palette = "jco", ggtheme = theme_classic())


clara.res_k8 <- clara(comp, 8, samples = 50, pamLike = TRUE,metric='euclidean')
fviz_cluster(clara.res, data = comp, stand = FALSE,
             ellipse = FALSE, show.clust.cent = FALSE,
             geom = "point",palette = "jco", ggtheme = theme_classic())
fviz_silhouette(clara.res_k8, palette = "jco", ggtheme = theme_classic())
```
